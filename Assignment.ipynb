```python
non_default = ap_data.loc[ap_data['TARGET'] == 0]
default = ap_data.loc[ap_data['TARGET'] == 1]
```

```python
#function for plotting different figures
def univariate_analysis(data, df, ndf, column):
    
    data = data
    non_dft = ndf
    dft = df
    col = column
    
    fig = plt.figure(figsize=(20,8))
    
    ax1 = plt.subplot(1,3,1)
    data[col].value_counts().plot.pie(autopct = "%1.0f%%" ,ax=ax1)
    plt.title('Plotting data for the column: '+ column)
    
    ax2 = plt.subplot(1,3,2)
    sns.countplot(x= column, hue = 'TARGET', data = data,palette= ['g','r'], ax = ax2)
    plt.xticks(rotation=90)
    plt.title('Graph of Target vs '+col+" in terms of count")
    plt.legend(['Non-Defaulter','Defaulter'])

    #c=['red', 'yellow', 'black', 'blue', 'orange']
    ax3 = plt.subplot(1,3,3)
    (pd.pivot_table(data=ap_data,index=col,values='TARGET')*100).plot.bar(ax=ax3)
    plt.title("Graph of TARGET vs "+col+" in terms of percentage")

    plt.legend(['Defaulter'])
    plt.tight_layout() 

    plt.show()
      
```

1.CODE_GENDER:- If the applicant is male

2.NAME_FAMILY_STATUS:- People having civil marriage or single

3.NAME_EDUCATION_TYPE:- People with Lower education 

4.NAME_INCOME_TYPE:-Clients who are either at Maternity leave OR Unemployed default a lot.
5.REGION_RATING_CLIENT:- Applicant who live in Rating 3 has highest defaults.

6.OCCUPATION_TYPE:- Low-skill Laborers, Drivers and Waiters/barmen staff, Security staff, Laborers and Cooking staff as the default rate is huge.

7.ORGANIZATION_TYPE:- Organizations with highest percent of loans not repaid are Transport: type 

8.DAYS_BIRTH:- Young people who are in age group of 20-40 as they have higher probability of defaulting

10.DAYS_EMPLOYED:- Applicant who have less than 5 years of employment have high default rate.

11.CNT_CHILDREN & CNT_FAM_MEMBERS:- Applicant who have children equal to or more than 9 default 100% and hence their applications are to be rejected.
